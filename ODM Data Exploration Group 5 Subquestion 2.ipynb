# Data ingestion

xlsx File from: 
https://www.england.nhs.uk/statistics/statistical-work-areas/covid-19-daily-deaths/


```python
import pandas as pd # Imports pandas library with alias pd for short
import matplotlib.pyplot as plt # Plotting library matplotlib
```

```python
# Reads from .xlsx file and saves the file in a variable

# Daily count
## Link changes with the date. This is the latest at the moment (30th of August 2020).
# (Plots wouldn't change if filter remains for 11-March to 30-July)
Daily_dec = pd.ExcelFile("https://www.england.nhs.uk/statistics/wp-content/uploads/sites/2/2020/08/COVID-19-total-announced-deaths-30-August-2020.xlsx")

# Hospital activity
# Hospital_act = pd.ExcelFile("Covid-Publication-13-08-2020 (hospital activity - Dataset 2).xlsx")
```

```python
# Reads the specified sheet (from the excel) in Daily_Dec
daily_3 = pd.read_excel(Daily_dec, sheet_name=2)  # Note that first worksheet is 0, not 1
daily_4 = pd.read_excel(Daily_dec, sheet_name=3)
daily_5 = pd.read_excel(Daily_dec, sheet_name=4)
daily_6 = pd.read_excel(Daily_dec, sheet_name=5)
# The variable's type is pandas DataFrame
display(daily_3.head(18))
# Note the data starts on row 14 (16 in the excel file, it counts from row 2 of the excel file)
```

### Getting the part of the spreadsheet with the data we are interested in

```python
# Dictionary variable with contents of each spreadsheet data
Content_dict = {}
Content_dict["daily_3"] = daily_3[daily_3.columns[2]].iloc[1]; print(Content_dict["daily_3"],"\n")
Content_dict["daily_4"] = daily_4[daily_3.columns[2]].iloc[1]; print(Content_dict["daily_4"],"\n")
Content_dict["daily_5"] = daily_5[daily_3.columns[2]].iloc[1]; print(Content_dict["daily_5"],"\n")
Content_dict["daily_6"] = daily_6[daily_3.columns[2]].iloc[1]; print(Content_dict["daily_6"],"\n")
```

```python
# Rows where the data values are (14 of our DataFrame)
# Assigns the area with data to the variable (daily_n = slice of daily_n)
daily_3 = daily_3[14:]
daily_4 = daily_4[14:]
daily_5 = daily_5[14:]
daily_6 = daily_6[14:]

display(daily_3.head(5))
```

### Cleaning data

```python
# Removing empty columns or rows
daily_3.dropna(how = "all", axis = 1, inplace=True)  # Removes columns (axis=1) where all values are False/Empty
daily_3.dropna(axis = 0, inplace=True)               # Removes rows    (axis=0) where some value is False/Empty (last is a 0)
daily_4.dropna(how = "all", axis = 1, inplace=True)
daily_4.dropna(axis = 0, inplace=True)
daily_5.dropna(how = "all", axis = 1, inplace=True)
daily_5.dropna(axis = 0, inplace=True)
daily_6.dropna(how = "all", axis = 1, inplace=True)
daily_6.dropna(axis = 0, inplace=True)
```

```python
def date_clean(df):
    '''Goes throug all columns of the row with index '14' (the row with dates)
    and transforms dates into strings (str) to remove the year, the hour:minute:second and empty spaces'''
    for col in df.columns:
        df[col].loc[14] = str(df[col].loc[14]).strip("00:00:00")
        df[col].loc[14] = str(df[col].loc[14]).strip("2020-")
        df[col].loc[14] = str(df[col].loc[14]).strip()

date_clean(daily_3)    # Calls the function for every DataFrame (with the spreadsheets)
date_clean(daily_4)
date_clean(daily_5)
date_clean(daily_6)

daily_4
```

```python
# Sets the column names with first row
daily_3.columns = daily_3.iloc[0]
# Delete first row (Title and dates)
daily_3 = daily_3.drop(14,axis = 0)

daily_4.columns = daily_4.iloc[0]
daily_4 = daily_4.drop(14,axis = 0)
daily_5.columns = daily_5.iloc[0]
daily_5 = daily_5.drop(14,axis = 0)
daily_6.columns = daily_6.iloc[0]
daily_6 = daily_6.drop(14,axis = 0)

daily_5
```

```python
# Sets new index with more appropiate row names contained in ("Name") etc. Deletes old columns' row name
daily_3 = daily_3.set_index("NHS England Region")
del daily_3.columns.name
daily_4 = daily_4.set_index("NHS England Region")
del daily_4.columns.name
daily_5 = daily_5.set_index("Age group (years)")
del daily_5.columns.name
daily_6 = daily_6.set_index("Name")
del daily_6.columns.name
daily_3
```

### Filtering the desired data (London area and 3-Mar to 31-Jul)

```python

daily_6 = daily_6[daily_6["NHS England Region"]=="London"].copy() # Makes a copy of the area-filtered Dataframe
daily_6.drop(["NHS England Region","Code"],axis=1,inplace=True)   # Removes the region 'Code' column
daily_5.drop("TBC",axis=0,inplace=True)  # Removes the To be confirmed from the age data (up to date all zeros)
```

```python
def timerange(df):
    ''' Function to filter date. Making lists of the column names (dates)
    searches for the desired column (date) index with the method .index(). Then removes the columns out of
    these indexes.'''
    start=list(df.columns).index("3-11")
    df.drop(df.columns[range(0,start)],axis=1,inplace=True)
    end=list(df.columns).index("7-31")
    awaiting=list(df.columns).index("Awaiting verification")
    df.drop(df.columns[range(end+1,awaiting)],axis=1,inplace=True)

timerange(daily_3)
timerange(daily_4)
timerange(daily_5)
timerange(daily_6)
daily_5
```

```python
print(daily_5.iloc[0][:].values.max(axis=0)) # Prints the max of the values on first row.
pd.to_numeric(daily_6.iloc[0]).idxmax()      # Gives the index or column name of the maximum value
```

# Retrieving the desired data and plotting

### Finding peaks and the average value for London trusts other than Imperial College

```python
# List of all London trusts
London_trusts = list(daily_6.index)
# Finds index of row with Imperial College data
Imperial_row = London_trusts.index("IMPERIAL COLLEGE HEALTHCARE NHS TRUST")
# List with numbers of the list of London trusts and removing the Imperial one
ind=list(range(len(London_trusts)))
ind.remove(Imperial_row)
# Creates a pandas Series with the mean value of all rows but the Imperial College one
newrow = daily_6.iloc[ind].mean()
newrow = pd.Series(newrow)
# Names the pd.Series and adds it at the end of the DataFrame
newrow.name = "AVERAGE OTHER TRUSTS"
daily_6 = daily_6.append(newrow,ignore_index=False)
daily_6.tail(3)
```

```python
# Creates a column to all the DataFrames (spreadsheets) to add the date of the peak (max value for the timeline)
daily_3["Peak day"]= 0
daily_4["Peak day"]= 0
daily_5["Peak day"]= 0
daily_6["Peak day"]= 0

def peak(df):
    '''It takes a DataFrame, and for every index (row) it searches for the column position (date)
    of the max value and gets also the value itself (it avoids last two columns 
    [Awaiting verification] and [Total]). Prints those values and saves the date on the new column "Peak day".'''
    for ind in df.index:
        date = pd.to_numeric(df.loc[ind][:-2]).idxmax()
        value = df.loc[ind][:-2].max()
        print(ind,date,value)
        df["Peak day"].loc[ind] = date
    print()
    
peak(daily_3)   # Calls to the function
peak(daily_4)   
peak(daily_5)   
peak(daily_6)   
```

```python
# Sorts values of "Peak day", retrieves the index of the rows sorted and makes a list.
# On that list looks for "IMPERIAL COLLEGE ..." and returns the position it has.
Imp_pos = list(daily_6.sort_values(by=["Peak day"]).index).index('IMPERIAL COLLEGE HEALTHCARE NHS TRUST')
Imp_pos
```

### Plotting values

```python
# Creates the figure 'frame' and plot space with certain size
fig, ax = plt.subplots(figsize=(15, 14))

# Data
trusts = daily_6.sort_values(by=["Peak day"]).index              # Row name of sorted trusts by 'Peak day' column
y_pos = daily_6.sort_values(by=["Peak day"])["Peak day"].values  # Values for each sorted row (date)

# Formating
color='b'
barlist = ax.barh(trusts,y_pos, align='center',color=color, alpha=0.5)  # Sets the format for each bar
ax.invert_yaxis()                                                       # labels read top-to-bottom
plt.tick_params(axis='y', labelsize=16)                   
plt.tick_params(axis='x', labelsize=16,rotation=45)        # Axis font size and rotates the date
# ax.grid(axis="x")     # Plots with grid (off)
barlist[Imp_pos].set_color('r')                   # Sets the bar corresponding to Imperial (Imp_pos) to red
ax.set_xlabel('Day',fontsize=16)                  # Font size of X axis label
# ax.set_ylabel('Trust')

# Sets title and saves the figure
ax.set_title('Day when the first peak was reached by trust (London area)',fontsize=20)
fig.savefig("Date of peak by trust"+".jpg", bbox_inches = 'tight')

plt.show()
```

```python
# For the peaks with line plots of the worksheets we define a function to plot any of them.

def lineplot(df,name):
    '''This function takes a DataFrame  and a name. The name is used in the plot title, to tell apart the two
    differently structured spreadsheets and for the file name of the saved plot figure'''
    
    fig, ax = plt.subplots(figsize=(15, 10))               # Figure size
    
    # Diferent behaviour depending on the spreadsheet (daily_3,4,5 or daily_6)
    if name == "Deaths per day by trust for London area":
        # Creates a dictionary for the two diferent colors depending on the row name
        dic = {"IMPERIAL COLLEGE HEALTHCARE NHS TRUST":"r", "AVERAGE OTHER TRUSTS":"b"}  
        for IMP,color in dic.items():           # Loops the dictionary for the row name and color                                            
            pd.to_numeric((df.loc[IMP][:-2])).plot(color=color,alpha=0.7) # Plots the line for every value
            date = pd.to_numeric(df.loc[IMP][:-2]).idxmax()               # Saves the index (date) of each peak
            date_sort = list(df.columns).index(date)                      # Gets the column position of the peak
            value = df.loc[IMP][:-2].max()                                # Gets the value of the peak
            # Creates annotations for each line, pointing to the peak (date_sort(column position),value)
            # and writes the row name (IMP) at a slightly displaced position (x,y) -> (x*1.1,y*1.1)
            ax.annotate(IMP, xy=(date_sort, value), xycoords="data", xytext=(date_sort*1.1,value*1.03),
                    textcoords='data',arrowprops=dict(arrowstyle="->",connectionstyle="arc3"),fontsize="13")         
            print(IMP,date,value)

    else:     # the other spreadsheets
        for ind in df.index:                                           # For each row
            pd.to_numeric((df.loc[ind][:-2])/df.loc[ind][-2]).plot()   # every value / total [Normalised]
            date = pd.to_numeric(df.loc[ind][:-2]).idxmax()            # analogous to the other plot but 
            date_sort = list(df.columns).index(date)
            value = df.loc[ind][:-2].max()/df.loc[ind][-2]             # with normalised values
            ax.annotate(ind, xy=(date_sort, value), xycoords="data", xytext=(date_sort*1.1,value*1.03),
                    textcoords='data',arrowprops=dict(arrowstyle="->",connectionstyle="arc3"),fontsize="13")          
            print(ind,date,value)
    # Creates the legend
    legend = plt.legend(loc = 'best', 
                    shadow=True,
                    prop={'weight':'roman','size':'large'})
                         
    # Label the axes and sets the ticks
    plt.xlabel('Day',fontsize="16")
    plt.ylabel("Deaths per day",fontsize="16")
    plt.tick_params(axis='y', labelsize=14)
    plt.tick_params(axis='x', labelsize=14,rotation=45)
    # Label the figure and saves the file
    plt.title(name,fontsize="22")
    fig.savefig(name+".jpg")
    plt.show()

# To improve: "Normalised" on title?, slightly bigger legend lines, more Xticks,
# line of each peak, rolling average to smoothen curves (and remove data mistakes by short delays on reporting)
```
